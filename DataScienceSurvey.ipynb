# Data Science Survey Results

#### In this projecct, Results of a survay on various data science topics are analyzed to create proper visualizations.

```python
# Import libraries.

import numpy as np
import pandas as pd
```

```python
# Download dataset for analysis.

survey = pd.read_csv('https://cocl.us/datascience_survey_data/Topic_Survey_Assignment.csv')

print('Data downloaded and read into a dataframe!')
```

```python
# Review data.

survey.head()
```

```python
# Rename columns.

survey.rename(columns={'Unnamed: 0':'Data Science Topics'}, inplace=True)
```

```python
# Set index to the dataset.

survey.set_index('Data Science Topics', inplace=True)
```

```python
survey.head()
```

```python
# Sort data for better understading.

survey.sort_values(by='Very interested', ascending=False, axis=0, inplace=True)
survey.head()
```

```python
# Normalize data

survey['Very interested'] = survey['Very interested']*100/2233
survey['Somewhat interested'] = survey['Somewhat interested']*100/2233
survey['Not interested'] = survey['Not interested']*100/2233
```

```python
survey.head()
```

```python
# Generate plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
ax = survey.plot(kind = 'bar',
                figsize = (20,8),
                width = 0.8,
                color = ('#5cb85c', '#5bc0de', '#d9534f'),
                fontsize = 14)
ax.set_title("Percentage of Respondents' Interest in Data Science", fontsize = 16)
ax.legend(fontsize = 14)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)


for p in ax.patches:
    ax.annotate("{:.2%}".format(p.get_height()),
                xy=(p.get_x()+0.01, p.get_height()+0.5))
```
